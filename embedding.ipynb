```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size = 10000
```

```python
## One hot representation 
one_hot_repr = [one_hot(words, voc_size) for words in sent]
## to tell at which index is the specific word located
one_hot_repr
```

```python
## word Embedding Representation

from tensorflow.keras.layers import Embedding
##from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## total number of words a sentence can have in a list
sent_length=8
## tells the one hot repr to to add the additional entries before the actual words in a sentence if needed to set it to 8 words for each sentence
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
## feature representation
dim = 10
```

```python
model=Sequential()
## Embedding have integrated Word2Vec and memorize this syntax
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile(optimizer='adam', loss='mse')
```

```python
# Build the model explicitly by providing an input shape(because if this is not done then the summary will show empty dataset)
model.build(input_shape=(None, sent_length))
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
